## **Linear Interpolation**

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

## **INTERPOLATING**

```python
x_new=np.linspace(1,5,10)
y_interpolate=np.interp(x_new,x,y)
y_interpolate
```

```python
plt.scatter(x_new,y_interpolate)
```

## **CUBIC INTERPOLATION**

```python
x=np.array([1,2,3,4,5])
y=np.array([2,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
f=interp1d(x,y,kind="cubic")
```

```python
x_new=np.linspace(1,5,10)
y_interpolate=f(x_new)
y_interpolate
```

```python
plt.scatter(x_new,y_interpolate)
```

## **POLYNOMIAL INTERPOLATION**

```python
a=np.array([1,2,3,4,5])
b=np.array([1,4,9,16,25])
```

```python
p=np.polyfit(a,b,2)
a_new=np.linspace(1,5,10)
b_interpolate=np.polyval(p,a_new)
```

```python
plt.scatter(a,b)
```

```python
plt.scatter(a_new,b_interpolate)
```

```python
p=np.polyfit(a,b,4)
a_new=np.linspace(1,5,10)
b_interpolate=np.polyval(p,a_new)
plt.scatter(a_new,b_interpolate)
```
