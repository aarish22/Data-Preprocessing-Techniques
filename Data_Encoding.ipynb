## **One Hot Encoding(Nominal Data)**

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({
    'hobbies':['singing','dancing','cooking','movies','painting','cycling']
})
```

```python
df
```

```python
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(df[['hobbies']])
```

```python
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

## **Ordinal Encoding**

## **Target Guided Ordinal Encoding**

```python
movie_ticket=pd.DataFrame({
    'movies_name':['Doctor Strange 2','Loki','Thor','Iron Man','Hulk','Hawkeye','Flash','Batman'],
    'ticket_cost':[344,233,456,213,323,433,345,467]
})
```

```python
movie_ticket
```

```python
movie_ticket.head()
```

```python
mean_price=movie_ticket.groupby('movies_name')['ticket_cost'].mean().to_dict()
```

```python
mean_price
```

```python
movie_ticket['movies_encoded']=movie_ticket['movies_name'].map(mean_price)
```

```python
movie_ticket['movies_encoded']
```

```python
movie_ticket
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
avg_price=tips.groupby('size')['total_bill'].mean().to_dict()
```

```python
avg_price
```

```python
tips['size_encoded']=tips['size'].map(avg_price)
```
