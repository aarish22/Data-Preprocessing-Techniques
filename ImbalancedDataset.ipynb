### Under_Sampling_Methods

```python
from sklearn.datasets import make_classification
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X , y = make_classification(n_samples = 1000,n_features = 2,n_redundant = 0, weights = [0.9] , random_state = 98, n_clusters_per_class=2)
```

```python
df = pd.DataFrame(X)
df['target'] = y
df
```

```python
df['target'].value_counts()
```

```python
class_0 = df[df['target'] == 0]
class_1 = df[df['target'] == 1]
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(class_0[0], class_0[1], c='blue', label='Class 0')
plt.scatter(class_1[0], class_1[1], c='red', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Two Classes')
plt.legend()
plt.show()
```

### We can see class 0 is much higher in numbers than class 1

### Random Under Sampling Method

```python
from imblearn.under_sampling import RandomUnderSampler   ## importing library
```

```python
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)
```

### After UnderSampling

```python
df_res = pd.DataFrame(X_res)
df_res['target'] = y_res

# Separate the classes
class_0_res = df_res[df_res['target'] == 0]
class_1_res = df_res[df_res['target'] == 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(class_0_res[0], class_0_res[1], c='blue', label='Class 0 (Resampled)')
plt.scatter(class_1_res[0], class_1_res[1], c='red', label='Class 1 (Resampled)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Resampled Classes')
plt.legend()
plt.show()
```

```python
df_res['target'].value_counts()
```

### Over-Sampling

```python
from sklearn.datasets import make_classification
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X1 , y1 = make_classification(n_samples = 1000,n_features = 2,n_redundant = 0, weights = [0.9] , random_state = 98, n_clusters_per_class=2)
```

```python
df = pd.DataFrame(X1)
df['target'] = y1
df
```

```python
class_0 = df[df['target'] == 0]
class_1 = df[df['target'] == 1]
```

```python
df['target'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(class_0[0], class_0[1], c='blue', label='Class 0')
plt.scatter(class_1[0], class_1[1], c='red', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Two Classes')
plt.legend()
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
X1_resampled, y1_resampled = ros.fit_resample(X1, y1)
```

### After OverSampling

```python
df_resampled = pd.DataFrame(X1_resampled)
df_resampled['target'] = y1_resampled

# Separate the classes
class_0_resampled = df_resampled[df_resampled['target'] == 0]
class_1_resampled = df_resampled[df_resampled['target'] == 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(class_0_resampled[0], class_0_resampled[1], c='blue', label='Class 0 (Resampled)')
plt.scatter(class_1_resampled[0], class_1_resampled[1], c='red', label='Class 1 (Resampled)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Resampled Classes')
plt.legend()
plt.show()
```

### We might not see much changes in the scatter_plot after oversampling because in oversampling the datapoints of minority class is duplicated and one red-dot may represent many data_points

## SMOTE Technique(OverSampling)

```python
from imblearn.over_sampling import SMOTE
```

```python
X1 , y1 = make_classification(n_samples = 1000,n_features = 2,n_redundant = 0, weights = [0.9] , random_state = 98, n_clusters_per_class=2)
df = pd.DataFrame(X1)
df['target'] = y1
df
```

```python
class_0 = df[df['target'] == 0]
class_1 = df[df['target'] == 1]
```

```python
df['target'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(class_0[0], class_0[1], c='blue', label='Class 0')
plt.scatter(class_1[0], class_1[1], c='red', label='Class 1')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Two Classes')
plt.legend()
plt.show()
```

```python
sm = SMOTE(random_state=42)
X1_res, y1_res = sm.fit_resample(X1, y1)
```

```python
df_res = pd.DataFrame(X1_res)
df_res['target'] = y1_res

# Separate the classes
class_0_res = df_res[df_res['target'] == 0]
class_1_res = df_res[df_res['target'] == 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(class_0_res[0], class_0_res[1], c='blue', label='Class 0 (Resampled)')
plt.scatter(class_1_res[0], class_1_res[1], c='red', label='Class 1 (Resampled)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Resampled Classes')
plt.legend()
plt.show()
```

### Smothing techinque is a type of over_sampling method that uses the concept of interpolation and KNN(for finding the nearest neighbours then interpolating point the new data points between the two data points)
