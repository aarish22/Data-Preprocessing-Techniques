### **Standardization**

```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
df['total_bill']
```

```python
total_bill=list(df['total_bill'])
total_bill
```

```python
import numpy as np
mean=np.mean(total_bill)
stdev=np.std(total_bill)
```

```python
mean,stdev
```

```python
normalized_data=[]
for i in total_bill:
  z_score=(i-mean)/stdev
  normalized_data.append(z_score)

normalized_data
```

```python
sns.kdeplot(total_bill)
```

```python
sns.kdeplot(normalized_data)
```

## **Using Scikit-Learn**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill','tip']])
```

```python
import pandas as pd
pd.DataFrame(scaler.transform(df[['total_bill','tip']]),columns=['total_bill','tips'])
```
